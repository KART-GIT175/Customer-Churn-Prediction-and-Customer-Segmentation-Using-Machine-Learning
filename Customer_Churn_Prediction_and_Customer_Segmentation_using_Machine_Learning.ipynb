```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report
```

```python
data = pd.read_csv('/content/WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.columns.unique()
```

```python
df.info()
```

```python
df.drop('customerID', axis=1, inplace=True)
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df['gender'] = df['gender'].map({'Female':0,'Male':1})
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['TotalCharges']=df['TotalCharges'].fillna(df['TotalCharges'].mean())
```

```python
df = df.astype(float)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
x = df.drop('Churn_Yes', axis=1)
y = df['Churn_Yes']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(x_train, y_train)
```

```python
ylog = log.predict(x_test_scaled)
ylog
```

```python
print(classification_report(y_test, ylog))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
yknn = knn.predict(x_test_scaled)
yknn
```

```python
print(classification_report(y_test, yknn))
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train,y_train)
```

```python
ynb = nb.predict(x_test_scaled)
```

```python
ynb
```

```python
print(classification_report(y_test, ynb))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
ydt = dt.predict(x_test_scaled)
```

```python
ydt
```

```python
print(classification_report(y_test, ydt))
```

```python
y_svm = svm.predict(x_test_scaled)
```

```python
y_svm
```

```python
print(classification_report(y_test, y_svm))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
y_rf = rf.predict(x_test_scaled)
```

```python
y_rf
```

```python
print(classification_report(y_test, y_rf))
```

```python
models = {
    "Logistic Regression": LogisticRegression(),
    "KNN": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier()
}
from sklearn.model_selection import cross_val_score

for name, model in models.items():

    scores = cross_val_score(
        model,
        x_train,
        y_train,
        cv=5,
        scoring='accuracy'
    )

    print(name)
    print("CV Scores:", scores)
    print("Mean CV Score:", scores.mean())
    print()

```

```python
from sklearn.ensemble import StackingClassifier
estimators = [
    ('knn', KNeighborsClassifier()),
    ('dt', DecisionTreeClassifier()),
    ('rf', RandomForestClassifier()),
    ('svc', SVC()),
    ('log', LogisticRegression()),
    ('gb', GradientBoostingClassifier()),
    ('ada', AdaBoostClassifier()),
    ('xgb', XGBClassifier()),
    ('nb', GaussianNB()),
    ('lr', LogisticRegression())
]
stack = StackingClassifier(estimators=estimators, final_estimator=KNeighborsClassifier())
stack.fit(x_train,y_train)
```

```python
y_stack = stack.predict(x_test_scaled)
```

```python
y_stack
```

```python
print(classification_report(y_test, y_stack))
```
